```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0.0, -0.4911, -0.2423, -0.2845]
y = [0.0, -0.0437, -0.0918, 0.7672]
mean_x = np.mean(x)
mean_y = np.mean(y)
# Shifting the centroid of the plot to the origin
x = x - mean_x
y = y - mean_y - 0.15
plt.scatter(x, y, marker='o')
# Make x and y axes have the same scale
plt.axis('equal')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
print(x)
print(y)
```
